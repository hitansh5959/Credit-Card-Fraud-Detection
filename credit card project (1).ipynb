##**COMPARISON BETWEEN LOGISTIC,SVC AND KNN**

```python
 # Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# Load synthetic dataset
X, y = make_classification(n_samples=1000, n_features=20, n_informative=2, n_redundant=10, random_state=42)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize classifiers
svm_classifier = SVC(kernel='linear')
knn_classifier = KNeighborsClassifier(n_neighbors=5)
logistic_classifier = LogisticRegression(max_iter=1000)

# Train classifiers
svm_classifier.fit(X_train, y_train)
knn_classifier.fit(X_train, y_train)
logistic_classifier.fit(X_train, y_train)

# Predict and calculate accuracy
svm_accuracy = accuracy_score(y_test, svm_classifier.predict(X_test))
knn_accuracy = accuracy_score(y_test, knn_classifier.predict(X_test))
logistic_accuracy = accuracy_score(y_test, logistic_classifier.predict(X_test))

# Data for bar graph
classifiers = ['SVM', 'KNN', 'Logistic Regression']
accuracies = [svm_accuracy, knn_accuracy, logistic_accuracy]

# Plot bar graph
plt.figure(figsize=(10, 6))
plt.bar(classifiers, accuracies, color=['blue', 'green', 'red'])
plt.xlabel('Classifier')
plt.ylabel('Accuracy')
plt.title('Comparison of Classifier Accuracies')
plt.ylim(0, 1)  # Assuming accuracy is between 0 and 1
for i in range(len(accuracies)):
    plt.text(i, accuracies[i] + 0.02, f'{accuracies[i]:.2f}', ha='center')
plt.show()
```
